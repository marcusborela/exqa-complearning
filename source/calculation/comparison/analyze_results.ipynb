```python
df_pivoted_metric = rastro_eval_qa._df_calculated_metric.pivot(index='cod_evaluation', columns='cod_metric', values='value')
df_pivoted_metric
```

```python
df_eval_pivoted = pd.merge(left=rastro_eval_qa.df_evaluation, right=df_pivoted_metric , left_on='cod', right_on='cod_evaluation', how='left')
df_eval_pivoted
```

## Resultados das avaliações


```python
df_eval_pivoted['Shots'] = df_eval_pivoted['cod_prompt_format'].astype('Int64') % 100 - 1
```

```python
df_eval_final=df_eval_pivoted[['cod', 'name_learning_method', 'ind_language', 'name_model',
       'num_max_answer_length', 'Shots', 'EM','EM@3','F1','F1@3']].sort_values(['F1@3'], ascending=False)
df_eval_final
```

```python
df_eval_final.to_csv('data/vw_evaluation.csv', index=False)
```

```python
df_times = rastro_eval_qa.df_evaluation[['cod', 'name_learning_method', 'ind_language', 'name_model', 'num_max_answer_length', 'cod_prompt_format', 
                                         'time_execution_total', 'time_execution_per_question']]
```

```python
df_times['Shots'] = df_times['cod_prompt_format'].astype('Int64') % 100 - 1
```

```python
df_time_total = df_times[['cod', 'name_learning_method', 'ind_language', 'name_model', 'num_max_answer_length', 'Shots', 'time_execution_total']].sort_values(['time_execution_total'], ascending=False)
df_time_total
```

```python
df_time_question = df_times[['cod', 'name_learning_method', 'ind_language', 'name_model', 'num_max_answer_length', 'Shots', 'time_execution_per_question']].sort_values(['time_execution_per_question'], ascending=False)
df_time_question
```

```python
df_time_total.to_csv('data/vw_duration_total.csv', index=False)
df_time_question.to_csv('data/vw_duration_question.csv', index=False)
```
